```python
movies = pd.read_csv("/kaggle/input/movie-rating-dataset/movie.csv")
movies.head()
```

```python
ratings = pd.read_csv("/kaggle/input/movie-rating-dataset/rating.csv")
ratings.head()
```

```python
df = ratings.merge(movies,how="left",on="movieId")
df.head()
```

```python
# remove below 1000 commented movies
df.groupby("movieId").agg({"userId":"count"})

```

```python
greater_than_1000_movie_counts = df.groupby("movieId").agg({"userId":"count"})
greater_than_1000_movie_counts = greater_than_1000_movie_counts[greater_than_1000_movie_counts["userId"]>1000]
greater_than_1000_movie_counts.sort_values("userId",ascending=True)
```

```python
df_filtered = df[df["movieId"].isin(greater_than_1000_movie_counts.index)]
df_filtered.shape
```

```python
df_pivot = df_filtered.pivot_table(values = "rating",index = "userId",columns ="title" )
df_pivot.head()
```

```python
#select random user
random_user = df_filtered["userId"].sample(1,random_state=45).values[0]
random_user
```

```python
random_user_df = df_filtered[df_filtered["userId"] == random_user]
random_user_df.head()
```

```python
random_user_df.shape
```

```python
movies_watched = random_user_df["title"].to_list()
```

```python
movies_watched_df = df_pivot[movies_watched]
movies_watched_df.head()
```

```python
user_movie_count = movies_watched_df.T.notna().sum()
user_movie_count
```

```python
type(user_movie_count)
```

```python
user_movie_count / len(movies_watched) > 0.60
```

```python
# %60 ından fazlasını izleyenleri sec
users_same_movies = user_movie_count[user_movie_count / len(movies_watched) > 0.60].index.to_list()
```

```python
len(users_same_movies)
```

```python
movies_watched_df_filtered = movies_watched_df.iloc[users_same_movies,:]
movies_watched_df_filtered.shape
```

```python
movies_watched_df_filtered.head()
```

```python
# add user to that list to calculate correlation
random_user_pivot = df_filtered[df_filtered["userId"]==random_user].pivot_table(values="rating",index="userId",columns ="title" )
random_user_pivot
```

```python
random_user_pivot = random_user_pivot[movies_watched]
random_user_pivot
```

```python
final_df = pd.concat([movies_watched_df_filtered,random_user_pivot])
final_df
```

```python
final_df.T.corr()
```

```python
corr_df = final_df.T.corr()
corr = corr_df.loc[random_user,:]
corr
# SORU : buradaki negative korelasyonlar ne anlama geliyor? benzerlik bulmak için korelasyondan başka yöntem var mı?
```

```python
corr[corr>0.65]
```

```python
similar_users = corr[corr>0.65]
similar_users.to_frame()
similar_users = similar_users.reset_index()
similar_users.columns = ["userId","corr"]
similar_users = similar_users[similar_users["userId"] != random_user]

similar_users
```

```python
similar_users.sort_values(by="corr",ascending=False)
```

```python
top_users = df_filtered[df_filtered["userId"].isin(similar_users["userId"])]
```

```python
top_users = top_users.merge(similar_users,how = "left",on = "userId")
top_users
```

```python
# Weighted
top_users["weighted_rating"] = top_users["corr"] * top_users["rating"]
top_users
```

```python
top_users.sort_values(by = "weighted_rating",ascending=False)
```

```python
recommendation_df = top_users.groupby("movieId").agg({"weighted_rating":"mean"})
recommendation_df
```

```python
type(recommendation_df)
```

```python
recommendation_df = recommendation_df[recommendation_df["weighted_rating"] > 3.5].sort_values(by = "weighted_rating",ascending=False)
recommendation_df
```
